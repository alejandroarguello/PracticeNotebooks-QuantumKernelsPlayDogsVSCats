```python
train_dir = "./data/datasetCatsVSDogs/train"

categorias = ['cats', 'dogs']

dataset = []

for categoria in categorias:
  path = os.path.join(train_dir, categoria) #ruta de la carpeta cats o la carpeta dogs
  label = categorias.index(categoria)

  i=0
  for img in os.listdir(path):
    if i<200:
      imgpath = os.path.join(path, img) #ruta de la imagen
      imagen = cv2.imread(imgpath,1) # el 0 indica que se cargue la imagen en escala de grises, un 1 indica en color
      try:
        imagen = cv2.resize( imagen, (128,128))
        imagen = np.array(imagen).flatten() #es como hacer ToTensor()
        #cv2_imshow(imagen) #muestra la imagen

        dataset.append([imagen, label])
      except:
        pass
    i += 1
    
```

```python
print(len(dataset))
```

```python
features = []
labels = []

for feature,label in dataset:
  features.append(feature) #almacena por un lado una lista con las features
  labels.append(label) #almacena por otro lado una lista con las labels de las imagenes


xtrain, xtest, ytrain, ytest = train_test_split(features, labels, test_size = 0.20, shuffle= True)
#separa la features (x) y las labels (y) y los grupos de train y test
```

**MODELO SVM**
